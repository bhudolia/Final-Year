```python
try:
    import urllib.request as urllib2
except ImportError:
    import urllib2
from bs4 import BeautifulSoup

```

```python
 articleURL = "https://www.imdb.com/title/tt7098658/reviews?ref_=tt_ql_3"
```

```python
page = urllib2.urlopen(articleURL).read().decode('utf8','ignore')
soup = BeautifulSoup(page,"lxml")
print(soup)
```

```python
soup.find('div')
```

```python
soup.find("body").get_text()
```

```python
for data in soup.findAll('div',{'class':'text show-more__control'}):
    print(data.text)
```

```python
alist = [] 
for data in soup.findAll('div',{'class':'text show-more__control'}):
    #print(data.text)
    alist.append(data.text)
print(alist[1])
```

```python
print(alist[5])
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sents = sent_tokenize(alist[5])
print(sents)
```

```python
word_sent = word_tokenize(alist[5].lower())
print(word_sent)
```

```python
 _stopwords = set(stopwords.words('english') + list(punctuation))
print(_stopwords)
```

```python
 word_sent =[word for word in word_sent if word not in _stopwords]
```

```python
print(word_sent)
```

```python
word_sent
```

```python
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
```

```python
print(freq)
```

```python
from heapq import nlargest 
nlargest(10, freq, key = freq.get)
```

```python
from collections import defaultdict 
ranking = defaultdict(int)
for i,sent in enumerate(sents):
    for w in word_tokenize(sent):
        if w in freq:
            ranking[i] += freq[w]
ranking
```

```python
sents_idx = nlargest(4, ranking, key=ranking.get)
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```
